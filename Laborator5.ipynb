```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
url_iris = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
names = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class']
iris_df = pd.read_csv(url_iris, names=names)
iris_df.head()
```

```python
#Din urmatorul grafic putem observa care sunt sepalele cu dimensiunile cele mai mari
iris_df.head(25).plot(kind='scatter',x='sepal_length',y='sepal_width',color='red')
plt.show()
```

```python
#In acest grafic putem observa petalele cu cele ma mari dimensiuni
iris_df.head(20).plot(kind='bar',x='petal_length',y='petal_width')
```

```python
#In acest grafic putem observa cum evolueaza dimensiunile sepaleleor pe o anumita clasa
#gca=get current axis
ax = plt.gca()

iris_df.plot(kind='line',x='class',y='sepal_length',ax=ax)
iris_df.plot(kind='line',x='class',y='sepal_width', color='red', ax=ax)

plt.show()
```

```python
iris_df.groupby(['class','sepal_width'])['sepal_length'].size().unstack().plot(kind='bar',stacked=True)
plt.show()
```

```python
iris_df.head(30).plot.area()
```
